Import pandas,Seaborn,matplotlib

```python
import pandas as pd
import seaborn as sns
sns.set(style='whitegrid')
import matplotlib.pyplot as plt
df=pd.read_csv('student_grades.csv')
df.head()
```

# Bar Plot
Student Count per Grade

```python
# Bar Plot: Number of students per grade
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='Grade', hue='Grade', palette='pastel', legend=False)
plt.title('Number of Students per Grade')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

 # Histogram –
 Distribution of Average Marks

```python
# Histogram: Distribution of average scores
plt.figure(figsize=(6, 4))
sns.histplot(df['Average'], bins=5, kde=True, color='skyblue')
plt.title('Distribution of Average Scores')
plt.xlabel('Average Score')
plt.ylabel('Number of Students')
plt.tight_layout()
plt.show()
```

# Box Plot –
Spread of Marks per Subject

```python
# Box Plot: Range of marks in each subject
plt.figure(figsize=(8, 5))
sns.boxplot(data=df[['Subject1', 'Subject2', 'Subject3']], palette='Set2')
plt.title('Spread of Marks per Subject')
plt.ylabel('Marks')
plt.tight_layout()
plt.show()
```

# Line Plot -
Total Scores by Student

```python
# Line Plot: Total marks by student (sorted)
plt.figure(figsize=(8, 4))
sorted_df = df.sort_values('Total')
plt.plot(sorted_df['Name'], sorted_df['Total'], marker='o', color='orange')
plt.title('Total Marks by Student')
plt.xlabel('Student')
plt.ylabel('Total Marks')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#  Pie Chart -
Grade Distribution

```python
# Pie Chart: Grade distribution
plt.figure(figsize=(5, 5))
df['Grade'].value_counts().plot.pie(autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Grade Distribution')
plt.ylabel('')  # Remove y-label for a clean look
plt.tight_layout()
plt.show()
```

# Heatmap -

Shows Relationship between: Subject1, Subject2, Subject3, Total, Average. Values close to 1 → strong positive relation & Values close to 0 → weak relation

```python
# Load your data
df = pd.read_csv('student_grades.csv')

# Select only numeric columns
numeric_df = df.select_dtypes(include=['number'])

# Create heatmap
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')

plt.title("Correlation Heatmap of Student Data")
plt.show()
```

# Observation -

```python
# Observation based on visualizations
print("📊 Observations:")

# Grade distribution
grade_counts = df['Grade'].value_counts()

# Print summary
for grade, count in grade_counts.items():
    print(f" - Grade {grade}: {count} students")

print("\n✅ Visual Insights:")
print("- The bar and pie charts show the majority of students fall into Grade B.")
print("- The histogram indicates most average scores are between 60 and 80.")
print("- Box plots reveal Subject 2 has slightly lower median scores compared to others.")
print("- The line plot shows variation in total scores, highlighting top and low performers.")
print("- The Heatmap shows strong correlation between Total and Average since both are derived values. Subjects also show moderate correlation indicating consistent student performance.")
```
